```python
# LinkedIn Profile Reviewer Pro

## Project Overview
This Python application analyzes a LinkedIn profile and provides feedback, scoring, and improvement suggestions.

### Features
- Profile scoring
- AI keyword detection
- GitHub portfolio check
- Project mention analysis
- Career objective verification
```

```python
print("=" * 50)
print("      LINKEDIN PROFILE REVIEWER PRO")
print("=" * 50)

headline = input("\nEnter Your LinkedIn Headline:\n")

about = input("\nEnter Your About Section:\n")

score = 0

strengths = []
suggestions = []

# Headline Check
if len(headline) >= 15:
    score += 20
    strengths.append("Good headline length")
else:
    suggestions.append("Write a longer headline")

# AI Keywords Check
keywords = ["python", "machine learning", "ai", "data science",
            "gen ai", "artificial intelligence"]

found = False

for word in keywords:
    if word in about.lower():
        found = True

if found:
    score += 20
    strengths.append("AI related skills mentioned")
else:
    suggestions.append("Add AI/ML skills")

# GitHub Check
if "github" in about.lower():
    score += 20
    strengths.append("GitHub portfolio mentioned")
else:
    suggestions.append("Add GitHub profile")

# Projects Check
if "project" in about.lower():
    score += 20
    strengths.append("Projects mentioned")
else:
    suggestions.append("Mention projects")

# Career Goal Check
if "engineer" in about.lower() or "career" in about.lower():
    score += 20
    strengths.append("Career objective found")
else:
    suggestions.append("Add career objective")

print("\n")
print("=" * 50)
print("             REVIEW REPORT")
print("=" * 50)

print("\nProfile Score:", score, "/100")

print("\nStrengths:")

if len(strengths) == 0:
    print("No strengths detected")
else:
    for item in strengths:
        print("✓", item)

print("\nSuggestions:")

if len(suggestions) == 0:
    print("Excellent Profile")
else:
    for item in suggestions:
        print("✗", item)

print("\nOverall Rating:")

if score >= 80:
    print("Outstanding LinkedIn Profile")
elif score >= 60:
    print("Good Profile - Needs Small Improvements")
elif score >= 40:
    print("Average Profile")
else:
    print("Needs Major Improvement")

print("\nThank You For Using LinkedIn Profile Reviewer Pro")
```
